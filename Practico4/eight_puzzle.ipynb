### Random Agent

Possible actions: Up=0, Right=1, Down=2, Left=3


```python
start_state_flatten = env.reset()
done = False
step_counter = 0
all_rewards = 0
env.render()

while not done:
    action = env.action_space.sample()
    obs, reward, done_env, _ = env.step(action)
    all_rewards += reward
    done = done_env
    env.render()
    step_counter += 1
    if (step_counter == 25):  #Early stopping
        done = True

print("Finished in", step_counter, "steps")
```
